# Graphing for College Algebra

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-1,10,999)

def f(x):
  return (1-x**3)**(1/2)

def g(x):
  return -1*(1-x**3)**(1/2)

y = f(x)
yg = g(x)

plt.plot(x,y)
plt.plot(x,yg)
```
